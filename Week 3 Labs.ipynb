## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
plt.figure(figsize=(5,3))
sns.boxplot(x=df.ConvertedComp, data=df)
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
firstq = df["ConvertedComp"].quantile(0.25)
thirdq = df["ConvertedComp"].quantile(0.75)
interquartilerange = thirdq - firstq
print(interquartilerange)
```

Find out the upper and lower bounds.


```python
upper=thirdq+(interquartilerange*1.5)
lower=firstq-(interquartilerange*1.5)

print("Upper:", upper)
print("Lower:", lower)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
(df["ConvertedComp"]<lower) | (df["ConvertedComp"]>upper)
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
less = (df["ConvertedComp"] < (lower))
less.value_counts()
```

```python
more = (df["ConvertedComp"] > (upper))
more.value_counts()
```

```python
df2 = df[~(df["ConvertedComp"] > (upper))]
```
